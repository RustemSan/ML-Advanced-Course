Split into train/val/test (60/20/20)

````python
X = df_data.iloc[:, 1:].values 
y = df_data['label'].values

X_tensor = torch.tensor(X, dtype=torch.float32)        
y_tensor = torch.tensor(y, dtype=torch.long)           

dataset = TensorDataset(X_tensor, y_tensor)

generator1 = torch.Generator().manual_seed(random_seed)

train_data, temp_data = random_split(dataset, [0.6, 0.4], generator=generator1)
val_data, test_data   = random_split(temp_data, [0.5, 0.5], generator=generator1)

print("Train set size:     ", len(train_data))
print("Validation set size:", len(val_data))
print("Test set size:      ", len(test_data))
````

# MLP

## Pipeline Sanity Check

**Objective:** Ensure the core pipeline runs without errors.

**Actions:**
- **Model:** MLP with one hidden layer (256 neurons), ReLU activation  
- **Optimizer:** Adam with learning rate = 0.001  
- **Training settings:** `batch_size = 64`, `num_epochs = 10`  
- **Metrics to log each epoch:**  
  - `train_loss`  
  - `val_loss`  
  - `train_acc`  
  - `val_acc`

Create DataLoaders for mini-batch training

````python
g = torch.Generator().manual_seed(random_seed)

batch_size = 64
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader   = DataLoader(val_data,   batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_data,  batch_size=batch_size, shuffle=False)
````

Define a simple MLP model

````python
class MLP(nn.Module):
    def __init__(self, input_size=1024, hidden_size=256, num_classes=10):
        super(MLP, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_size, hidden_size),
            nn.ReLU(),               # activation function
            nn.Linear(hidden_size, num_classes)
        )
    def forward(self, x):
        return self.model(x)

model     = MLP()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
````

Mini-batch training loop (10 epochs)

````python
num_epochs = 10
for epoch in range(1, num_epochs+1):
    model.train()
    running_loss, correct, total = 0.0, 0, 0
    
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss    = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * labels.size(0)
        preds = outputs.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total   += labels.size(0)
    
    train_loss = running_loss / total
    train_acc  = correct / total
    
    # Validation
    model.eval()
    val_correct, val_total = 0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            preds   = outputs.argmax(dim=1)
            val_correct += (preds == labels).sum().item()
            val_total   += labels.size(0)
    val_acc = val_correct / val_total
    
    print(f"Epoch {epoch}/{num_epochs} | "
          f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
          f"Val Acc: {val_acc:.4f}")
````

Batch size = 64 is workable, but we may try smaller (32) or larger (128) to see if the plateau moves higher.

We’ll now proceed to next step: Batch size & epochs exploration, to find best parameters where Val acc improves beyond ~0.80.

 ## Batch Size & Epochs Exploration

**Objective:** determine the best `batch_size` and `num_epochs` settings.

**Fixed configuration:**
- Model: MLP(256, ReLU)  
- Optimizer: Adam(lr=0.001)  
- No regularization  

**Grid to explore:**
- `batch_size` ∈ {32, 64, 128}  
- `num_epochs` ∈ {10, 15, 20, 50, 100}  

**Metric:** final validation accuracy  

````python
batch_sizes   = [32, 64, 128]
epoch_options = [10, 15, 20, 50, 100]

results = []

for batch_size in batch_sizes:
    for num_epochs in epoch_options:
        train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
        val_loader   = DataLoader(val_data,   batch_size=batch_size)

        torch.manual_seed(random_seed)
        model     = MLP()
        criterion = nn.CrossEntropyLoss()
        optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

        for _ in range(num_epochs):
            model.train()
            for inputs, labels in train_loader:
                optimizer.zero_grad()
                outputs = model(inputs)
                loss    = criterion(outputs, labels)
                loss.backward()
                optimizer.step()

        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                outputs = model(inputs)
                preds   = outputs.argmax(dim=1)
                correct += (preds == labels).sum().item()
                total   += labels.size(0)
        val_acc = correct / total

        results.append({
            'batch_size': batch_size,
            'num_epochs': num_epochs,
            'val_acc':    val_acc
        })

df_results = pd.DataFrame(results)
print("\nValidation Accuracy (batch_size vs. num_epochs):")
print(df_results.pivot(index='batch_size', columns='num_epochs', values='val_acc'))
````

Visualize how batch size and number of epochs affect validation accuracy using a heatmap

````python
pivot_table = df_results.pivot(index='batch_size', columns='num_epochs', values='val_acc')

plt.figure(figsize=(8, 6))
sns.heatmap(
    pivot_table,
    annot=True,        
    fmt=".4f",         
    cmap="YlGnBu",      
    cbar_kws={'label': 'Validation Accuracy'}
)
plt.title("Validation Accuracy for Batch Size vs. Num Epochs")
plt.xlabel("Number of Epochs")
plt.ylabel("Batch Size")
plt.yticks(rotation=0)
plt.show()
````

The heatmap clearly shows that increasing the number of training epochs yields substantial gains in validation accuracy across all batch sizes. With only 10–20 epochs, accuracy hovers around 0.77–0.82, but at 50 epochs it jumps into the 0.83–0.88 range, and by 100 epochs we consistently exceed 0.90. Batch size has a smaller effect: sizes 32 and 64 perform comparably (peak ≈0.91 at 32×100 epochs and ≈0.907 at 64×100), while 128×100 still reaches a strong 0.904.

Key takeaway: To balance training time and performance, using 50 epochs with a batch size of 64 gives a solid ~0.88 validation accuracy without the lengthy 100-epoch run.

 ## Activation Function Comparison

**Objective:** Evaluate how different activation functions affect validation accuracy.

**Fixed settings:**
- Model architecture: MLP(1024→256→10)  
- Optimizer: Adam(lr=0.001)  
- `batch_size = 64`  
- `num_epochs = 50`  

**Activations to test:**
- `ReLU`  
- `LeakyReLU(0.01)`  
- `SELU`  
- `Tanh`  

**Metric:** final validation accuracy  

````python
activation_functions = {
    'ReLU':      nn.ReLU(),
    'LeakyReLU': nn.LeakyReLU(0.01),
    'SELU':      nn.SELU(),
    'Tanh':      nn.Tanh() # wanted to try this activation function
}

batch_size    = 64
num_epochs    = 50
learning_rate = 0.001

train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader   = DataLoader(val_data,   batch_size=batch_size)

# 4) Container for loss curves
loss_curves = {}

for name, act_fn in activation_functions.items():
    torch.manual_seed(random_seed)
    
    model = nn.Sequential(
        nn.Flatten(),
        nn.Linear(1024, 256),
        act_fn,
        nn.Linear(256, 10)
    )
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    
    # Record average train loss each epoch
    epoch_losses = []
    
    for epoch in range(1, num_epochs+1):
        model.train()
        running_loss = 0.0
        total        = 0
        
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss    = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * labels.size(0)
            total        += labels.size(0)
        
        epoch_losses.append(running_loss / total)
    
    loss_curves[name] = epoch_losses

# 5) Plot all curves on one chart
plt.figure(figsize=(8,6))
for name, losses in loss_curves.items():
    plt.plot(range(1, num_epochs+1), losses, label=name)

plt.yscale('log')              # optional: log-scale to spread out early epochs
plt.xlabel('Epoch')
plt.ylabel('Average Training Loss')
plt.title('Training Loss vs. Epoch for Different Activations')
plt.legend()
plt.grid(True, which='both', ls='--', alpha=0.5)
plt.show()
````

The loss curves reveal clear differences in how quickly and how far each activation drives training:

**SELU (green)** converges the fastest and reaches the lowest loss by epoch 50, indicating strong gradient flow and very effective feature scaling.

**ReLU (blue)** closely follows SELU after the initial few epochs, steadily decreasing loss and ending at the second-best value.

**LeakyReLU (orange)** is slightly slower than plain ReLU but still outperforms Tanh, showing better training dynamics than a hard zero gradient on negatives.

**Tanh (red)** converges the slowest and plateaus early around a much higher loss (~0.8–1.0), which explains its lower final accuracy.

Key takeaway: SELU gives the fastest and deepest descent, so it’s the best choice here, with ReLU as a strong runner-up

# Optimizer Comparison

**Objective:** Compare how different optimizers affect convergence and final validation accuracy.

**Fixed settings:**
- Model architecture: MLP(1024→256→10) with **SELU** activation  
- `batch_size = 64`  
- `num_epochs = 50`  
- `learning_rate = 0.001`  

**Optimizers to test:**
- `torch.optim.Adagrad`  
- `torch.optim.RMSprop`  
- `torch.optim.Adam`  

**Metric:** final validation accuracy  

````python
optimizers = {
    'Adagrad': torch.optim.Adagrad,
    'RMSprop': torch.optim.RMSprop,
    'Adam':    torch.optim.Adam,
    'AdamW':   torch.optim.AdamW
}

batch_size    = 64
num_epochs    = 50
learning_rate = 0.001

results_opt = []

for name, OptimizerClass in optimizers.items():
    train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(val_data,   batch_size=batch_size)
    
    torch.manual_seed(random_seed)
    model     = nn.Sequential(
        nn.Flatten(),
        nn.Linear(1024, 256),
        nn.SELU(),
        nn.Linear(256, 10)
    )
    criterion = nn.CrossEntropyLoss()
    optimizer = OptimizerClass(model.parameters(), lr=learning_rate)
    
    for epoch in range(num_epochs):
        model.train()
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss    = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
    
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            preds   = outputs.argmax(dim=1)
            correct += (preds == labels).sum().item()
            total   += labels.size(0)
    val_acc = correct / total
    
    print(f"{name:8s} Validation Accuracy: {val_acc:.4f}")
    results_opt.append({'optimizer': name, 'val_acc': val_acc})

df_opt = pd.DataFrame(results_opt).set_index('optimizer')
````

Visualization

````python
opt_names = df_opt.index.tolist()
opt_scores = df_opt['val_acc'].values

plt.figure(figsize=(6, 4))
bars = plt.bar(opt_names, opt_scores, color=['#A6CEE3','#B2DF8A','#FB9A99'])  # pastel colors

for bar, score in zip(bars, opt_scores):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.01,
             f"{score:.4f}", ha='center', va='bottom')

plt.ylim(0, 1)
plt.title("Validation Accuracy by Optimizer")
plt.xlabel("Optimizer")
plt.ylabel("Validation Accuracy")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

The bar chart makes it clear that AdamW outperforms all other optimizers here, achieving a validation accuracy of 0.9525. Adam comes in second at 0.8892, a solid improvement over Adagrad (≈0.8221), while RMSprop trails at 0.7836.

Key takeaway: AdamW’s built-in weight decay gives it an edge in generalization on this MLP, so I will adopt AdamW for all remaining experiments.

 # Regularization Techniques

**Objective:** Assess the impact of Dropout and Batch Normalization on validation accuracy and overfitting.

**Fixed settings:**
- Model architecture: `MLP(1024→256→10)` with **SELU** activation  
- Optimizer: **AdamW** (since it performed best) with `lr=0.001`  
- batch_size = 64  
- num_epochs = 50

**Configurations to test:**
1. **No regularization**  
2. **Dropout** after the hidden layer (`p=0.5`)  
3. **BatchNorm** before the activation  
4. **Dropout + BatchNorm**  

**Metrics:**  
- Final validation accuracy  
- Train vs. validation accuracy gap (to gauge overfitting)  

````python
configs = {
    'none':      {},  
    'dropout':   {'dropout': True,  'batchnorm': False},
    'batchnorm': {'dropout': False, 'batchnorm': True},
    'both':      {'dropout': True,  'batchnorm': True},
}

batch_size    = 64
num_epochs    = 50
learning_rate = 0.001

results_reg = []

for name, opts in configs.items():
    torch.manual_seed(random_seed)
    # build the network dynamically
    layers = [ nn.Flatten(), nn.Linear(1024, 256) ]
    
    if opts.get('batchnorm', False):
        layers.append(nn.BatchNorm1d(256))
    layers.append(nn.SELU())
    
    if opts.get('dropout', False):
        layers.append(nn.Dropout(p=0.5))
    
    layers.append(nn.Linear(256, 10))
    model = nn.Sequential(*layers)
    
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate)
    
    train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(val_data,   batch_size=batch_size)
    
    # training loop
    for epoch in range(num_epochs):
        model.train()
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss    = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
    
    # compute train accuracy
    model.eval()
    train_correct, train_total = 0, 0
    with torch.no_grad():
        for inputs, labels in train_loader:
            preds = model(inputs).argmax(dim=1)
            train_correct += (preds == labels).sum().item()
            train_total   += labels.size(0)
    train_acc = train_correct / train_total
    
    # compute val accuracy
    val_correct, val_total = 0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            preds = model(inputs).argmax(dim=1)
            val_correct += (preds == labels).sum().item()
            val_total   += labels.size(0)
    val_acc = val_correct / val_total
    
    results_reg.append({
        'config': name,
        'train_acc': train_acc,
        'val_acc':   val_acc,
        'gap':       train_acc - val_acc
    })

# Summarize
df_reg = pd.DataFrame(results_reg).set_index('config')
print(df_reg)
````

 # Hidden Layer Size & Learning Rate Tuning

**Objective:** Find the best combination of hidden layer width and learning rate for our final MLP.

**Fixed settings:**
- Architecture: MLP(1024 → *hidden_size* → 10)  
- Activation: SELU  
- Regularization: BatchNorm + Dropout(p=0.5)  
- Optimizer: AdamW  
- batch_size = 64  
- num_epochs = 50  

**Grid to explore:**
- `hidden_size` ∈ {128, 256, 512}  
- `learning_rate` ∈ {0.01, 0.005, 0.001, 0.0005, 0.0001}   

````python
import time
from itertools import product

hidden_sizes   = [128, 192, 256, 320]
learning_rates = [0.01, 0.005, 0.001, 0.0005, 0.0001]
num_epochs     = 50
batch_size     = 64

results_tune = []

# Start timing
start_time = time.time()

# progress bar
total_configs = len(hidden_sizes) * len(learning_rates)
for hidden_size, lr in tqdm(
    product(hidden_sizes, learning_rates),
    total=total_configs,
    desc="Grid Search",
    unit="config"
):
    torch.manual_seed(random_seed)

    # Build model with BatchNorm + Dropout
    model = nn.Sequential(
        nn.Flatten(),
        nn.Linear(1024, hidden_size),
        nn.BatchNorm1d(hidden_size),
        nn.SELU(),
        nn.Dropout(p=0.5),
        nn.Linear(hidden_size, 10)
    )
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr)

    train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(val_data,   batch_size=batch_size)

    model.train()
    for _ in range(num_epochs):
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss    = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

    # Validation
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            preds   = model(inputs).argmax(dim=1)
            correct += (preds == labels).sum().item()
            total   += labels.size(0)
    val_acc = correct / total

    results_tune.append({
        'hidden_size':   hidden_size,
        'learning_rate': lr,
        'val_acc':       val_acc
    })

# End timing
elapsed = time.time() - start_time
print(f"Total grid search time: {elapsed:.2f} seconds")

# Summarize results
df_tune = pd.DataFrame(results_tune)
print(df_tune.pivot(index='hidden_size', columns='learning_rate', values='val_acc'))
````


- The highest validation accuracy (≈0.9579) was achieved with **hidden_size=320** and **learning_rate=0.0005**.  
- Very close performance was also observed for:
  - `hidden_size=256`, `lr=0.0005` (≈0.9563)  
  - `hidden_size=128`, `lr=0.0005` (≈0.9566)  
- At a lower learning rate (0.0001), accuracy increases with larger hidden layers—from 0.9206 up to 0.9417—but still falls short of the optimal combinations.  
- At higher learning rates (≥0.005), performance drops sharply, especially with larger hidden layers, indicating an overly aggressive update step.

Creating CNN whose constructor accepts all hyperparameters for further manipulations

````python
from typing import Callable, Optional, Dict

class CNNParam:
    def __init__(self,
                 # Model spec (required)
                 conv_filters:   list[int],
                 activation_fn:  nn.Module,
                 dropout_conv:   float,
                 dropout_fc:     float,
                 use_bn:         bool,

                 # Training spec (required)
                 optimizer_fn:   Callable,
                 optimizer_params: Dict,

                 # Defaults
                 fc_size:        int   = 128,
                 num_classes:    int   = 10,
                 scheduler_fn:   Optional[Callable] = None,
                 scheduler_params: Optional[Dict]   = None,
                 criterion:      nn.Module         = nn.CrossEntropyLoss(),

                 # Data spec
                 train_ds        = None,
                 val_ds          = None,
                 batch_size:     int   = 64,
                 num_epochs:     int   = 10,
                 rd_seed:    int   = 42):
        torch.manual_seed(rd_seed)

        # 1) Build convolutional blocks
        layers, in_ch = [], 1
        for f in conv_filters:
            layers.append(nn.Conv2d(in_ch, f, kernel_size=3, padding=1))
            if use_bn:
                layers.append(nn.BatchNorm2d(f))
            layers.extend([
                activation_fn,
                nn.MaxPool2d(2, 2),
                nn.Dropout2d(dropout_conv)
            ])
            in_ch = f
        self.features = nn.Sequential(*layers)

        # 2) Build fully-connected head
        final_spatial = 32 // (2 ** len(conv_filters))
        feat_dim      = in_ch * final_spatial * final_spatial
        self.fc1      = nn.Linear(feat_dim, fc_size)
        self.drop_fc  = nn.Dropout(dropout_fc)
        self.fc2      = nn.Linear(fc_size, num_classes)

        # 3) Assemble model, loss, optimizer, scheduler
        self.criterion = criterion
        self.model     = nn.Sequential(
            self.features,
            nn.Flatten(),
            self.fc1,
            activation_fn,
            self.drop_fc,
            self.fc2
        )
        self.optimizer = optimizer_fn(self.model.parameters(), **optimizer_params)
        self.scheduler = (
            scheduler_fn(self.optimizer, **scheduler_params)
            if scheduler_fn is not None else None
        )

        # 4) Data loaders
        self.train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
        self.val_loader   = DataLoader(val_ds,   batch_size=batch_size)
        self.num_epochs   = num_epochs

    def evaluate(self, loader):                       # ── helper to see train-val gap, could be turn on or off
        self.model.eval()
        loss_tot = corr = samp = 0
        with torch.no_grad():
            for x, y in loader:
                logits = self.model(x)
                loss_tot += self.criterion(logits, y).item() * y.size(0)
                corr += (logits.argmax(1) == y).sum().item()
                samp += y.size(0)
        return loss_tot / samp, corr / samp
    
    def run(self, verbose: bool = False):
        """
        Train & validate the CNN.
        If verbose=True, prints per-epoch metrics.
        Returns: best_val_acc, history dict.
        """
        best_val_acc = 0.0
        history = {'train_loss': [], 'val_acc': []}

        for epoch in range(1, self.num_epochs + 1):
            # Training phase
            self.model.train()
            total_loss, total_samples = 0.0, 0
            for x, y in self.train_loader:
                self.optimizer.zero_grad()
                logits = self.model(x)
                loss   = self.criterion(logits, y)
                loss.backward()
                self.optimizer.step()
                total_loss    += loss.item() * y.size(0)
                total_samples += y.size(0)
            avg_loss = total_loss / total_samples
            history['train_loss'].append(avg_loss)

            # Validation phase
            self.model.eval()
            correct, total = 0, 0
            with torch.no_grad():
                for x, y in self.val_loader:
                    preds = self.model(x).argmax(dim=1)
                    correct += (preds == y).sum().item()
                    total   += y.size(0)
            val_acc = correct / total
            history['val_acc'].append(val_acc)

            # Scheduler step if provided
            if self.scheduler:
                try:
                    self.scheduler.step(val_acc)
                except TypeError:
                    self.scheduler.step()

            best_val_acc = max(best_val_acc, val_acc)
            if verbose:                                       # plotting accuracy after each epoch also could be turn on or off
                print(f"Epoch {epoch}/{self.num_epochs}  "
                      f"train_loss={avg_loss:.4f}  val_acc={val_acc:.4f}")

        return best_val_acc, history
````

# Regularization Comparison

Keeping the selected architecture, activation, and optimizer fixed, we now sweep across key regularization techniques to control overfitting:     

 - `BatchNorm: on vs. off in each convolutional block`
 - `Conv‐block Dropout: p ∈ [0.0, 0.2, 0.5]`
 - `Fully‐connected Dropout: p ∈ [0.0, 0.2, 0.5]`
   
For each combination, we train for 50 epochs and record the validation accuracy.

````python
reg_grid = list(product(
    [False, True],       # use_bn
    [0.0, 0.2, 0.5],     # drop_conv
    [0.0, 0.2, 0.5]      # drop_fc
))
total_configs = len(reg_grid)  

regs = []
# Wrap the loop in tqdm for time/ETA reporting
for use_bn, drop_conv, drop_fc in tqdm(
    reg_grid,
    total=total_configs,
    desc="Reg Sweep",
    unit="config"
):
    exp = CNNParam(
        conv_filters     = [16, 32],
        activation_fn    = best_act,
        dropout_conv     = drop_conv,
        dropout_fc       = drop_fc,
        use_bn           = use_bn,
        fc_size          = 128,
        num_classes      = 10,
        optimizer_fn     = best_opt_fn,
        optimizer_params = best_opt_params,
        train_ds         = train_cnn_ds,
        val_ds           = val_cnn_ds,
        batch_size       = 64,
        num_epochs       = 50,
        rd_seed      = random_seed
    )
    val_acc, history = exp.run(verbose=False)
    regs.append({
        'use_bn':    use_bn,
        'drop_conv': drop_conv,
        'drop_fc':   drop_fc,
        'val_acc':   val_acc
    })

# Pivot & heatmap (unchanged)
df_regs = pd.DataFrame(regs)
pivot = df_regs.pivot_table(
    index=['use_bn', 'drop_conv'],
    columns='drop_fc',
    values='val_acc'
)

plt.figure(figsize=(8, 6))
sns.heatmap(pivot, annot=True, fmt=".4f", cmap="viridis", linewidths=0.5, linecolor="white")
plt.title("Validation Accuracy: BatchNorm & Dropout Sweep (Adam, SELU)")
plt.xlabel("FC Dropout Rate")
plt.ylabel("Use BatchNorm / Conv Dropout Rate")
plt.show()
````

From the architecture sweep bar chart above, we observe:

The 2‐block (16→32) and 2‐block (32→64) models achieved the highest validation accuracies (≈99.1% and 99.07%, respectively), suggesting that two convolutional blocks are sufficient for this dataset.

Moving to a 3‐block architecture (16→32→64) causes a small drop (≈94.7%), and (32→64→128) actually recovers to ≈98.7%, indicating that depth and channel sizing both matter.

The deepest 4‐block (16→32→64→128) model underperforms (≈88.9%), likely due to overfitting or vanishing gradients.

Conclusion:
A relatively shallow CNN with two convolutional stages (either [16→32] or [32→64]) gives the best generalization here. Adding more blocks without adjusting other hyperparameters (e.g. learning rate, regularization) does not improve—and can even hurt—validation performance.

# Learning‐Rate Scheduler Comparison
 Now that we have fixed our best convolutional architecture, activation, optimizer
 and regularization settings, we will sweep over different learning-rate schedulers
 to see which one yields the highest validation accuracy:
   - ReduceLROnPlateau (drop LR when val_acc plateaus)
   - CosineAnnealingLR (cosine decay over epochs)
   - OneCycleLR (one-cycle policy with rising then falling LR)

````python
schedulers = {
    'ReduceLROnPlateau': lambda opt: torch.optim.lr_scheduler.ReduceLROnPlateau(
                                  opt, mode='max', factor=0.5, patience=5),
    'CosineAnnealingLR': lambda opt: torch.optim.lr_scheduler.CosineAnnealingLR(
                                  opt, T_max=50, eta_min=1e-5),
    'OneCycleLR':        lambda opt: torch.optim.lr_scheduler.OneCycleLR(
                                  opt, max_lr=1e-3,
                                       total_steps=50*len(train_loader))
}

results_sched = []
for name, make_sched in schedulers.items():
    exp = CNNParam(
        conv_filters     = best_conv_filters,    
        activation_fn    = best_act,             
        dropout_conv     = best_dropout_conv,    
        dropout_fc       = best_dropout_fc,      
        use_bn           = best_use_bn,          

        optimizer_fn     = best_opt_fn,          
        optimizer_params = best_opt_params,      

        scheduler_fn     = make_sched,
        scheduler_params = {},                   

        train_ds         = train_cnn_ds,         
        val_ds           = val_cnn_ds,
        batch_size       = 64,
        num_epochs       = 50,
        rd_seed      = random_seed
    )
    val_acc, _ = exp.run(verbose=False)
    results_sched.append({'scheduler': name, 'val_acc': val_acc})
    print(f"{name:<20} → val_acc = {val_acc:.4f}")

````

Plotting 

````python
df_sched = pd.DataFrame(results_sched).set_index('scheduler').sort_values('val_acc')

plt.figure(figsize=(8, 4))
sns.set_theme(style="whitegrid")

# Horizontal barplot
ax = sns.barplot(
    x='val_acc',
    y=df_sched.index,
    data=df_sched.reset_index(),
    palette='viridis',
    orient='h'
)

for i, (val) in enumerate(df_sched['val_acc']):
    ax.text(
        val + 0.002,           
        i,                     
        f"{val:.4f}", 
        va='center'
    )

# Tidy up the axes
ax.set_xlim(0.0, 1.0)
ax.set_xlabel("Validation Accuracy")
ax.set_ylabel("LR Scheduler")
ax.set_title("Validation Accuracy by Learning-Rate Scheduler")

plt.tight_layout()
plt.show()
````

From the horizontal‐bar comparison we see:

 - CosineAnnealingLR achieved the highest validation accuracy (≈ 0.9915), edging out ReduceLROnPlateau(≈ 0.9913).

 - OneCycleLR underperformed slightly (≈ 0.8786), likely because its cyclic schedule didn’t align well with our relatively small dataset and fixed epoch count.

Conclusion: We’ll adopt CosineAnnealingLR as our final LR scheduler.
